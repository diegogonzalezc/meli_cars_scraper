```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.float_format','{:,.2f}'.format)
```

```python
df_antioquia=pd.read_csv('Datos_obtenidos/antioquia.csv') # creación dataframes
df_risaralda=pd.read_csv('Datos_obtenidos/risaralda.csv')
df_caldas=pd.read_csv('Datos_obtenidos/caldas.csv')
df_quindio=pd.read_csv('Datos_obtenidos/quindio.csv')
list_df=[df_antioquia,df_risaralda,df_caldas,df_quindio]
```

```python
df_quindio.columns
```

```python
df_total=pd.DataFrame(columns=['Unnamed: 0', 'Marca', 'Modelo', 'Año', 'Color', 'Tipo de combustible',
       'Puertas', 'Transmisión', 'Motor', 'Tipo de carrocería', 'Kilómetros',
       'Precio', 'Link', 'Ubicación'])
for i in list_df: #se unen todos los dataframes para crear un dataframe total
       df_total=pd.concat([df_total,i])
df_total.drop(columns='Unnamed: 0', inplace=True)
df_total
```

```python
df_total.info() # se analiza la información que se obtiene
```

```python
df_total_clean=df_total.copy()
```

```python
df_total_clean['Precio']=df_total_clean['Precio'].str.replace('.','', regex=True) # a partir de aca se edita la información para ser utilizada, cambiamos tipos de datos y eliminamos datos que no sirvan.
df_total_clean['Kilómetros']=df_total_clean['Kilómetros'].str.replace(' km','', regex=True)
df_total_clean = df_total_clean.astype({"Año": int, "Kilómetros": float,'Precio':float})# transofrmación de datos a enteros
```

```python
df_total_clean
```

```python
df_total_clean['Tipo de combustible'].unique() # reviso valore unicos de Tipo de combustible para entender que pasa con esta columna.
```

```python
sin_color=(df_total_clean['Tipo de combustible']=='2') | (df_total_clean['Tipo de combustible']=='4') | (df_total_clean['Tipo de combustible']=='5') | (df_total_clean['Tipo de combustible']=='3') 
#al revisar con este filtro vimos que no habia color en los carros, como para mi es un item importante eliminare esta columna y los filtraré más adelante
```

```python
df_final_an=df_total_clean[~sin_color]
df_final_an['Tipo de combustible'].unique() #Ya no hay valores distintos a Tipo de combustible.
```

```python
df_final_an['Color'].unique()# todos son colores, no hay problema.
```

```python
df_final_an['Puertas'].unique()
```

```python
# a continuación haré todos los filtros segun mis necesidades.
precio=df_final_an['Precio']<=55000000
kilometraje=df_final_an['Kilómetros']<=65000
modelo=df_final_an['Año']>=2014
df_final_an=df_final_an[precio & kilometraje & modelo]
```

```python
df_final_an
```

```python
df_final_an['Marca'].unique()
```

```python
df_final_an['Marca'].value_counts()
```

```python
marcas_no_interesantes=(df_final_an['Marca']=='DFM/DFSK') | (df_final_an['Marca']=='Chevrolet Joy HB MT 2023') | (df_final_an['Marca']=='Dodge') | (df_final_an['Marca']=='Mahindra') | (df_final_an['Marca']=='Changhe')
 
```

```python
descartables=df_final_an[marcas_no_interesantes] # revisando estos carros no son interesantes para mi por lo tanto los descartaré
descartables
```

```python
df_final_an=df_final_an[~marcas_no_interesantes]
```

```python
df_final_an['Color'].unique()
#Se calificarán de 1 a 5 teniendo en cuenta la información que encontré en los enlaces que pondré al final, esta información es constante según las busquedas que he hecho.
#5 tendra el valor más alto de ahí para abajo irán disminuyendo.
```

```python
df_final_an['Color_cal']=df_final_an.Color.map({'Marrón':0, 'Azul':0, 'Rojo':0.75, 'Gris':2.2, 'Plateado':2.2, 'Blanco':5, 'Negro':1.75,
       'Dorado':0, 'Naranja':0})
df_final_an
```

```python
df_final_an['Marca'].unique() 
#Busqué las marcas de carros más confiables en 2022, según J.D. Power y la calificación va de 0 a 5 teniendo en cuenta la información allí contenida.
```

```python
df_final_an['Marca_cal']=df_final_an.Marca.map({'Chevrolet':4.8, 'Kia':4.9, 'Renault':3.9, 'Nissan':5, 'Volkswagen':2.7, 'Suzuki':2.7,
       'Fiat':2.7, 'Mazda':3.9, 'Ford':4.4, 'Hyundai':4.8})
df_final_an
```

```python
df_final_an['Ventas_cal']=df_final_an.Marca.map({'Chevrolet':4.9, 'Kia':2, 'Renault':5, 'Nissan':1.1, 'Volkswagen':0.7, 'Suzuki':2.3,
       'Fiat':0.05, 'Mazda':2.4, 'Ford':0.42, 'Hyundai':0.42})
df_final_an
# califcación seún que tan comercial son los vehiculos
```

```python
df_final_an['Precio_cal']=df_final_an['Precio'].apply(lambda x: 5 if x <= 40000000 else (4 if x > 40000000 and x < 44000000 else (3 if x > 44000000 and x < 48000000 else (2 if x > 48000000 and x < 52000000 else 1))))
# se califican de acuerdo al precio en un rango que va hasta los 65 millones
```

```python
df_final_an['Año_cal']=df_final_an['Año'].apply(lambda x: 5 if x == 2022 else (4 if x >= 2019 and x < 2021 else (3 if x >= 2017 and x < 2019 else (2 if x >= 2015 and x < 2017 else 1))))
# la calificacín de acuerdo a que tan nuevo o que tan viejo es el vehiculo en un rango de 2014  a2022
```

```python
df_final_an['Kilometraje_cal']=df_final_an['Kilómetros'].apply(lambda x: 5 if x <= 20000 else (4 if x > 20000 and x < 32000 else (3 if x > 32000 and x < 44000 else (2 if x > 44000 and x < 55000 else 1))))
# calificamos de acuerdo a que tantos kilometros tiene el carro
```

```python
df_final_an['Calificación']=df_final_an['Color_cal']*0.15+df_final_an['Marca_cal']*0.25+df_final_an['Ventas_cal']*0.1+df_final_an['Precio_cal']*0.2+df_final_an['Año_cal']*0.1+df_final_an['Kilometraje_cal']*0.2
df_final_an # para calificarlos se hará un ponderado poniendo el peso de cada una de las carácteristicas del carro
```

```python
df_final_an=df_final_an.sort_values('Calificación',ascending=False)
df_final_an.head(20)
```

```python
df_final_an=df_final_an[(df_final_an['Modelo'] != 'Kwid') & (df_final_an['Modelo']!='Spark GT') & (df_final_an['Modelo']!='Spark') & (df_final_an['Modelo']!='Twizy')] # retiro las referencias que no me llaman la atención
df_final_an=df_final_an.reset_index(drop=True)
df_final_an.head(20)
```
